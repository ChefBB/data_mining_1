```python
def preprocess_endyear(df):
    copy = df
    endYear = copy.loc[(copy['titleType'] == 'tvMiniSeries') | (copy['titleType'] == 'tvSeries'), 'endYear']

    copy['endYear'] = copy['startYear']
    copy.loc[(df['titleType'] == 'tvMiniSeries') | (copy['titleType'] == 'tvSeries'), 'endYear'] = endYear
    return copy['endYear']

df['endYear'] = preprocess_endyear(df)
```
