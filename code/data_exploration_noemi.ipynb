Confrontare diverse variabili

Raggruppo per rating e titleType -> vedo proporzione di title type per ogni rating

```python
grouped_ratings_type = train_df.groupby(['rating',"titleType"]).size().unstack(fill_value=0)

grouped_ratings_type.plot(kind="bar", stacked=True)
plt.savefig('plots/adult.png')

```

Raggruppo per titleType e rating -> vedo proporzione di rating per titletype

```python
grouped_type_ratings = train_df.groupby(["titleType",'rating']).size().unstack(fill_value=0)

grouped_type_ratings.plot(kind="bar", stacked=False)
plt.savefig('plots/grouped_type_ratings.png')
```
